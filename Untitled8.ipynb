```python
import numpy as np
from scipy.integrate import trapz, quad

def prob(x,lambd,N):
    return N*lambd*np.exp(-lambd*x)*(1-np.exp(-lambd*x))**(N-1)

def first_moment(x,lambd,N):
    return x*prob(x,lambd,N)
```

```python
import  matplotlib.pylab as plt

lambd = 3.0
MS=[1,2,3,4,4.5] #1.0*2**i for i in xrange(5)]
NMAX=[400,4000,80000,1000000,4000000]
dx=0.1
d={}
for m_s, Nmax  in zip(MS,NMAX):
    print m_s
    xsum = 0.0
    xsum = np.zeros(Nmax)
    
    x=np.arange(0,m_s,dx)
    
    for N in xrange(1,Nmax):
        xmean = trapz(first_moment(x,lambd,N), x=x, dx=dx)/trapz(prob(x,lambd,N), x=x, dx=dx)
    #    xmean = quad(first_moment,0,m_s)[0]/quad(prob,0,m_s)[0]
        xsum[ N-1 ] = xmean * (1-np.exp(-lambd*m_s))**N * np.exp(-lambd*m_s)
    d[m_s]=np.cumsum(xsum)[-1]
    
    plt.plot(np.cumsum(xsum))
    plt.show()
```

```python
def fit(x,a,b):
    return a*x+b

# plt.xscale('log')
# plt.yscale('log')
y=np.array([d[keys] for keys in d])
x=np.array([keys for keys in d])

z=np.polyfit(x, y,deg=1)

plt.scatter(x,y)
plt.plot(x,fit(x,*z))

plt.title('$m_{as}=%3.2f m_{ms}%3.2f$'%(z[0],z[1]))

plt.show()
```
